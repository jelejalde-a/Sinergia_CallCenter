El porcentaje de comisión varía según el tipo de servicio y valor, en busca optimizar los ingresos del call center sin afectar la rentabilidad de la empresa tercerizada.
comisión = f(valor del servicio, dificultad de venta, volumen esperado)

Servicios de bajo valor (15.000 - 25.000): 12-15% de comisión para el call center, ya que pueden ser más difíciles de vender, y generar mayor esfuerzo

Servicios de valor medio (30.000 - 60.000): 8-11% de comisión, equilibrio entre valor y esfuerzo

Servicios de alto valor (100.000+): 5.5-7% de comisión, menor porcentaje pero mayor ganancia absoluta

```python
from collections import Counter

HOY          = date(2026, 3, 25)
FECHA_INICIO = datetime(2026, 3,  1)
FECHA_FIN    = datetime(2026, 3, 25, 23, 59, 59)


MAX_DIAS_VACACIONES = 15
MAX_DIAS_LICENCIA   = 30

TURNOS_DEPTO = {
    'D4001': ['T002','T003'],
    'D4002': ['T006','T007','T008'],
    'D4003': ['T004','T005'],
    'D4004': ['T001'],
    'D4005': ['T001'],
}

NOMBRES_DEPTO = {
    'D4001': 'Ventas',
    'D4002': 'Soporte Técnico',
    'D4003': 'Servicio al Cliente',
    'D4004': 'Telemarketing',
    'D4005': 'Facturación y Cobranza',
}

cursor.execute("""
CREATE TABLE programacion (
    id_agente    VARCHAR(15)  NOT NULL,
    fecha_turno  DATE         NOT NULL
                 COMMENT 'Fecha en que el agente realiza el turno',
    id_turno     VARCHAR(10)  NOT NULL,
    asistencia   ENUM('asistido','falta','justificada','incapacidad','descanso')
                 DEFAULT 'asistido',
    PRIMARY KEY  (id_agente, fecha_turno),
    FOREIGN KEY  (id_agente) REFERENCES agente(id_agente),
    FOREIGN KEY  (id_turno)  REFERENCES turno(id_turno)
) ENGINE=InnoDB DEFAULT CHARSET=utf8mb4 COLLATE=utf8mb4_unicode_ci
""")

cursor.execute("""
    SELECT a.id_agente, a.id_departamento, a.id_cargo, a.estado
    FROM   agente a
    WHERE  a.estado != 'retirado'
    ORDER  BY a.id_departamento, a.id_cargo
""")
agentes_prog = cursor.fetchall()

print(f"\n{'='*65}")
print(f"  AGENTES A PROGRAMAR: {len(agentes_prog)}")
print(f"{'='*65}")
estados_count = Counter(row[3] for row in agentes_prog)
for est, n in sorted(estados_count.items()):
    print(f"    {est:<15}: {n}")
print()

def turno_rotativo(id_agente, fecha, turnos):
    agente_num = sum(ord(c) for c in id_agente) % len(turnos)
    semana     = fecha.isocalendar()[1]
    ciclo      = (semana // 2) % len(turnos)
    return turnos[(agente_num + ciclo) % len(turnos)]

def es_descanso_rotativo(id_agente, fecha):
    dia_del_anio  = fecha.timetuple().tm_yday
    offset_agente = sum(ord(c) for c in id_agente) % 7
    return (dia_del_anio + offset_agente) % 7 == 0

def asistencia_agente_activo():
    r = random.random()
    if r < 0.95:   return 'asistido'
    elif r < 0.98: return 'falta'
    else:          return 'justificada'

def calcular_turno(id_agente, departamento, fecha_dt):
    turnos  = TURNOS_DEPTO[departamento]
    dia_sem = fecha_dt.weekday()

    if departamento == 'D4001':
        if dia_sem == 6: return turnos[0], True
        turno = 'T003' if dia_sem == 5 else 'T002'
        return turno, False

    elif departamento == 'D4002':
        es_desc = es_descanso_rotativo(id_agente, fecha_dt)
        turno   = turno_rotativo(id_agente, fecha_dt, turnos) if not es_desc else turnos[0]
        return turno, es_desc

    elif departamento == 'D4003':
        es_desc = es_descanso_rotativo(id_agente, fecha_dt)
        turno   = ('T004' if sum(ord(c) for c in id_agente) % 2 == 0
                   else 'T005') if not es_desc else turnos[0]
        return turno, es_desc

    elif departamento in ('D4004', 'D4005'):
        if dia_sem in (5, 6): return turnos[0], True
        return turnos[0], False

    return turnos[0], False

# =============================================================================
# GENERACIÓN
# =============================================================================

programaciones = []
stats          = {}

for id_agente, departamento, cargo, estado in agentes_prog:

    stats[id_agente] = {
        'departamento':        departamento,
        'estado':              estado,
        'laborales':           0,
        'descansos':           0,
        'asistido':            0,
        'falta':               0,
        'justificada':         0,
        'incapacidad':         0,
        'dias_novedad_usados': 0,
    }

    cupo_novedad = (MAX_DIAS_VACACIONES if estado == 'vacaciones'
                    else MAX_DIAS_LICENCIA if estado == 'licencia'
                    else 0)
    dias_novedad_consumidos = 0
    fecha_actual = FECHA_INICIO

    while fecha_actual <= FECHA_FIN:
        turno_id, es_descanso = calcular_turno(id_agente, departamento, fecha_actual)

        if estado == 'incapacidad':
            asistencia = 'incapacidad'

        elif estado in ('vacaciones', 'licencia'):
            if es_descanso:
                asistencia = 'descanso'
            elif dias_novedad_consumidos < cupo_novedad:
                asistencia = 'justificada'
                dias_novedad_consumidos += 1
            else:
                asistencia = asistencia_agente_activo()

        else:
            asistencia = 'descanso' if es_descanso else asistencia_agente_activo()

        if asistencia == 'descanso':
            stats[id_agente]['descansos'] += 1
        else:
            stats[id_agente]['laborales'] += 1
            stats[id_agente][asistencia]  += 1

        stats[id_agente]['dias_novedad_usados'] = dias_novedad_consumidos

        programaciones.append((
            id_agente,
            fecha_actual.date(),
            turno_id,
            asistencia,
        ))
        fecha_actual += timedelta(days=1)

# =============================================================================
# INSERCIÓN
# =============================================================================

errores = 0
for p in programaciones:
    try:
        cursor.execute("""
            INSERT INTO programacion (id_agente, fecha_turno, id_turno, asistencia)
            VALUES (%s, %s, %s, %s)
        """, p)
    except Exception as e:
        errores += 1
        if errores <= 5:
            print(f"  x Error: {e} | agente={p[0]} fecha={p[1]}")
conexion.commit()

# =============================================================================
# ESTADÍSTICAS POR DEPARTAMENTO
# =============================================================================

print("\n" + "="*60)
print("  ESTADÍSTICAS POR DEPARTAMENTO — MARZO 2026")
print("="*60)

depto_stats = {}
for id_agente, datos in stats.items():
    d = datos['departamento']
    if d not in depto_stats:
        depto_stats[d] = {
            'agentes':0, 'activo':0, 'vacaciones':0,
            'licencia':0, 'incapacidad':0,
            'laborales':0, 'descansos':0,
            'asistido':0, 'falta':0,
            'justificada':0, 'incapacidad_dias':0,
        }
    ds = depto_stats[d]
    ds['agentes']          += 1
    ds[datos['estado']]     = ds.get(datos['estado'], 0) + 1
    ds['laborales']        += datos['laborales']
    ds['descansos']        += datos['descansos']
    ds['asistido']         += datos['asistido']
    ds['falta']            += datos['falta']
    ds['justificada']      += datos['justificada']
    ds['incapacidad_dias'] += datos['incapacidad']

for depto in sorted(depto_stats.keys()):
    ds       = depto_stats[depto]
    n_ag     = ds['agentes']
    prom_lab = ds['laborales']  / n_ag if n_ag else 0
    pct_asis = (ds['asistido']  / ds['laborales'] * 100) if ds['laborales'] else 0
    pct_falta= (ds['falta']     / ds['laborales'] * 100) if ds['laborales'] else 0

    print(f"\n  {'─'*54}")
    print(f"  {NOMBRES_DEPTO[depto]}  [{depto}]")
    print(f"  {'─'*54}")
    print(f"    Agentes programados        : {n_ag:>5}")
    print(f"      · activo                 : {ds.get('activo',0):>5}")
    print(f"      · vacaciones             : {ds.get('vacaciones',0):>5}")
    print(f"      · licencia               : {ds.get('licencia',0):>5}")
    print(f"      · incapacidad            : {ds.get('incapacidad',0):>5}")
    print(f"    Días laborales totales     : {ds['laborales']:>5}  (prom {prom_lab:.1f}/agente)")
    print(f"    Días de descanso totales   : {ds['descansos']:>5}")
    print(f"    Asistencia                 : {ds['asistido']:>5}  ({pct_asis:.1f}%)")
    print(f"    Faltas                     : {ds['falta']:>5}  ({pct_falta:.1f}%)")
    print(f"    Justificadas (nov/otros)   : {ds['justificada']:>5}")
    print(f"    Días incapacidad           : {ds['incapacidad_dias']:>5}")

print(f"\n{'='*60}")
print(f"  TOTAL REGISTROS PROGRAMACIÓN   : {len(programaciones):,}")
print(f"  ERRORES DE INSERCIÓN           : {errores}")
print(f"{'='*60}")
```

```python
cursor.execute("""
CREATE TABLE llamada (
    id_llamada          VARCHAR(15) PRIMARY KEY,
    id_agente           VARCHAR(15) NOT NULL,
    id_motivo           VARCHAR(10) NOT NULL,
    tipo_llamada        ENUM('entrante','saliente') NOT NULL,
    num_contacto        VARCHAR(20) NOT NULL,
    fecha_hora          DATETIME    NOT NULL,
    tiempo_espera       INT         COMMENT 'Segundos de espera (solo entrantes)',
    duracion            INT         COMMENT 'Duración en segundos',
    estado              ENUM('atendida','abandonada','transferida','no_contacto') NOT NULL,
    id_agente_destino   VARCHAR(15) COMMENT 'Agente al que se transfirió',
    resultado_gestion   ENUM(
        'venta_cerrada','ticket_resuelto','caso_resuelto',
        'retencion_efectiva','caso_concretado',
        'sin_resultado','no_aplica'
    ) NOT NULL DEFAULT 'no_aplica',
    satisfaccion        TINYINT UNSIGNED NULL,
    observaciones       TEXT,
    CONSTRAINT chk_satisfaccion_rango
        CHECK (satisfaccion IS NULL OR satisfaccion BETWEEN 1 AND 5),
    CONSTRAINT chk_tiempo_espera_tipo
        CHECK (
            (tipo_llamada = 'saliente' AND tiempo_espera IS NULL)
            OR (tipo_llamada = 'entrante')
        ),
    CONSTRAINT chk_duracion_estado
        CHECK (
            (estado IN ('abandonada','no_contacto') AND duracion IS NULL)
            OR (estado IN ('atendida','transferida'))
        ),
    FOREIGN KEY (id_agente)         REFERENCES agente(id_agente),
    FOREIGN KEY (id_agente_destino) REFERENCES agente(id_agente),
    FOREIGN KEY (id_motivo)         REFERENCES motivo_llamada(id_motivo),
    INDEX idx_fecha     (fecha_hora),
    INDEX idx_contacto  (num_contacto),
    INDEX idx_agente    (id_agente),
    INDEX idx_tipo      (tipo_llamada),
    INDEX idx_estado    (estado),
    INDEX idx_resultado (resultado_gestion)
) ENGINE=InnoDB DEFAULT CHARSET=utf8mb4 COLLATE=utf8mb4_unicode_ci
""")

# =============================================================================
# REFERENCIAS
# =============================================================================

cursor.execute("""
    SELECT a.id_agente, a.id_departamento, c.categoria
    FROM   agente a
    JOIN   cargo  c ON a.id_cargo = c.id_cargo
    WHERE  a.estado = 'activo'
""")
agentes_info = {
    row[0]: {'depto': row[1], 'nivel': row[2]}
    for row in cursor.fetchall()
}

cursor.execute("""
    SELECT p.id_agente, p.fecha_turno, t.hora_inicio, t.hora_fin
    FROM   programacion p
    JOIN   turno        t ON p.id_turno  = t.id_turno
    JOIN   agente       a ON p.id_agente = a.id_agente
    WHERE  p.fecha_turno BETWEEN '2026-03-01' AND '2026-03-25'
      AND  p.asistencia = 'asistido'
      AND  a.estado     = 'activo'
""")
programacion_index = {}
for id_ag, fecha, h_ini, h_fin in cursor.fetchall():
    if id_ag not in agentes_info:
        continue
    depto = agentes_info[id_ag]['depto']
    nivel = agentes_info[id_ag]['nivel']
    programacion_index.setdefault((fecha, depto), []).append(
        (id_ag, h_ini, h_fin, nivel)
    )

cursor.execute("""
    SELECT id_motivo, aplica_a, departamento_responsable
    FROM   motivo_llamada WHERE activo = 'si'
""")
motivos_por_depto = {}
for id_motivo, aplica_a, depto in cursor.fetchall():
    tipos = (['inbound','outbound'] if aplica_a == 'ambos' else [aplica_a])
    for t in tipos:
        motivos_por_depto.setdefault((depto, t), []).append(id_motivo)

# =============================================================================
# TABLAS DE LÓGICA
# =============================================================================

TIPO_A_BD = {'entrante': 'inbound', 'saliente': 'outbound'}

MOTIVO_A_TIPO_CASO = {
    'IN2-ST001': 'soporte_tecnico', 'IN2-ST002': 'soporte_tecnico',
    'IN2-ST003': 'soporte_tecnico', 'IN2-ST004': 'soporte_tecnico',
    'IN2-ST005': 'soporte_tecnico', 'OUT1-S001': 'soporte_tecnico',
    'OUT1-S002': 'soporte_tecnico', 'OUT1-S003': 'soporte_tecnico',
    'IN1-SC004': 'queja',           'IN1-SC005': 'reclamo',
    'IN1-SC006': 'solicitud',       'IN1-SC001': 'consulta',
    'IN1-SC002': 'consulta',        'IN1-SC003': 'consulta',
    'IN3-V001':  'consulta',        'IN3-V002':  'consulta',
    'IN3-V003':  'consulta',        'IN3-V004':  'consulta',
    'IN4-F001':  'consulta',        'IN4-F002':  'reclamo',
    'IN4-F003':  'consulta',        'IN4-F004':  'solicitud',
    'IN4-F005':  'solicitud',       'OUT4-F001': 'consulta',
    'OUT4-F002': 'reclamo',         'OUT4-F003': 'reclamo',
    'OUT4-F004': 'reclamo',         'OUT4-F005': 'solicitud',
    'OUT5-T001': 'consulta',        'OUT5-T002': 'consulta',
    'OUT5-T003': 'consulta',        'OUT5-T004': 'consulta',
    'OUT5-T005': 'consulta',        'OUT5-T006': 'consulta',
    'OUT5-T007': 'consulta',        'OUT3-V001': 'consulta',
    'OUT3-V002': 'consulta',        'OUT3-V003': 'consulta',
    'OUT3-V004': 'consulta',        'OUT3-V005': 'consulta',
    'NC006-SC3': 'consulta',
}

RUTAS_TRANSFERENCIA = {
    'D4001': ['D4003', 'D4005'],
    'D4002': ['D4003', 'D4001'],
    'D4003': ['D4002', 'D4005'],
    'D4004': ['D4003', 'D4001'],
    'D4005': ['D4003', 'D4002'],
}

POOL_A = list({
    f"3{random.randint(100_000_000, 199_999_999)}" for _ in range(4_000)
})[:2_500]
POOL_B = list({
    f"3{random.randint(100_000_000, 199_999_999)}" for _ in range(4_000)
} - set(POOL_A))[:2_000]

PCT_POOL_A = {
    'D4001': 0.30, 'D4002': 0.90,
    'D4003': 0.85, 'D4004': 0.70, 'D4005': 0.95,
}

def telefono_depto(depto):
    return (random.choice(POOL_A) if random.random() < PCT_POOL_A[depto]
            else random.choice(POOL_B))

VENTANA_HORARIA = {
    'D4001': (8,  18), 'D4002': (0,  23),
    'D4003': (6,  22), 'D4004': (8,  18), 'D4005': (8,  18),
}
DIAS_LABORABLES = {
    'D4001': [0,1,2,3,4,5], 'D4002': [0,1,2,3,4,5,6],
    'D4003': [0,1,2,3,4,5,6], 'D4004': [0,1,2,3,4], 'D4005': [0,1,2,3,4],
}
VOLUMEN_BASE = {
    'D4001': {'lun':660, 'mar_jue':555, 'vie':510, 'sab':270, 'dom':0},
    'D4002': {'lun':1020,'mar_jue':840, 'vie':780, 'sab':600, 'dom':450},
    'D4003': {'lun':1110,'mar_jue':930, 'vie':870, 'sab':660, 'dom':495},
    'D4004': {'lun':600, 'mar_jue':510, 'vie':450, 'sab':0,   'dom':0},
    'D4005': {'lun':510, 'mar_jue':435, 'vie':390, 'sab':0,   'dom':0},
}
MIX_LLAMADAS = {
    'D4001': {'entrante':0.55, 'saliente':0.45},
    'D4002': {'entrante':0.90, 'saliente':0.10},
    'D4003': {'entrante':0.85, 'saliente':0.15},
    'D4004': {'entrante':0.10, 'saliente':0.90},
    'D4005': {'entrante':0.40, 'saliente':0.60},
}
ESTADO_INBOUND = {
    'D4001': {'atendida':0.72, 'abandonada':0.18, 'transferida':0.10},
    'D4002': {'atendida':0.70, 'abandonada':0.20, 'transferida':0.10},
    'D4003': {'atendida':0.68, 'abandonada':0.22, 'transferida':0.10},
    'D4005': {'atendida':0.74, 'abandonada':0.16, 'transferida':0.10},
}
ESTADO_IB_DEF  = {'atendida':0.70, 'abandonada':0.20, 'transferida':0.10}
ESTADO_OUTBOUND = {
    'D4001': {'atendida':0.58, 'no_contacto':0.42},
    'D4004': {'atendida':0.48, 'no_contacto':0.52},
    'D4005': {'atendida':0.55, 'no_contacto':0.45},
}
ESTADO_OB_DEF = {'atendida':0.55, 'no_contacto':0.45}

# tasa 0.10 → ~1,215 ventas en marzo
# de esas ~100 serán clientes nuevos registrados en cliente
TASA_RESOLUCION = {
    'D4001': 0.10,
    'D4002': 0.68,
    'D4003': 0.65,
    'D4004': 0.15,
    'D4005': 0.35,
}
AJUSTE_NIVEL = {
    'junior':-0.10, 'senior':0.00, 'coordinador':+0.08,
    'supervisor':+0.10, 'lider':+0.15,
}
RESULTADO_DEPTO = {
    'D4001': 'venta_cerrada',    'D4002': 'ticket_resuelto',
    'D4003': 'caso_resuelto',    'D4004': 'retencion_efectiva',
    'D4005': 'caso_concretado',
}
OBS = {
    'abandonada': [
        "Cliente colgó durante el tiempo de espera",
        "Sin agente disponible — llamada no atendida",
        "Tiempo de espera superó umbral aceptable",
        "Alta demanda en la franja horaria",
        "Cliente desistió antes de ser atendido",
    ],
    'no_contacto': [
        "Buzón de voz — sin respuesta",
        "Línea ocupada, se reprogramará",
        "No contesta en los intentos del turno",
        "Número fuera de servicio — verificar en CRM",
        "Llamada rechazada por el cliente",
    ],
    'transferida': [
        "Transferida al departamento correspondiente",
        "Escalada a nivel superior por complejidad",
        "Redirigida al área especializada",
        "Transferida a supervisor por solicitud del cliente",
    ],
    'D4001_exito': [
        "Venta cerrada — activación en proceso",
        "Cliente aceptó el plan ofrecido",
        "Cierre con descuento de fidelización aplicado",
        "Venta concretada con upgrade incluido",
    ],
    'D4002_exito': [
        "Problema resuelto durante la llamada",
        "Ticket cerrado — falla corregida de forma remota",
        "Servicio restablecido — cliente confirmó",
        "Configuración ajustada exitosamente",
    ],
    'D4003_exito': [
        "PQR gestionada y cerrada en la misma llamada",
        "Consulta resuelta satisfactoriamente",
        "Solicitud procesada correctamente",
        "Reclamo atendido y solución comunicada",
    ],
    'D4004_exito': [
        "Cliente retenido con oferta personalizada",
        "Encuesta NPS completada — puntuación registrada",
        "Cliente confirmó continuidad del servicio",
        "Retención exitosa — beneficio de fidelización aplicado",
    ],
    'D4005_exito': [
        "Acuerdo de pago formalizado exitosamente",
        "Aclaración de factura realizada — cliente conforme",
        "Compromiso de pago obtenido",
        "Certificado tributario en generación",
    ],
    'sin_resultado': [
        "Cliente no interesado en la oferta",
        "Requiere tiempo para evaluar — seguimiento programado",
        "Caso derivado a supervisor por complejidad",
        "Cliente solicita información escrita antes de decidir",
        "Se brindó información sin concretar gestión",
    ],
}

# =============================================================================
# FUNCIONES
# =============================================================================

def generar_observacion(estado, resultado, depto):
    if estado == 'abandonada':
        return None if random.random() > 0.45 else random.choice(OBS['abandonada'])
    if estado == 'no_contacto':
        return None if random.random() > 0.45 else random.choice(OBS['no_contacto'])
    if estado == 'transferida':
        return None if random.random() > 0.80 else random.choice(OBS['transferida'])
    if resultado not in ('sin_resultado', 'no_aplica'):
        return random.choice(OBS.get(f'{depto}_exito', OBS['sin_resultado']))
    return None if random.random() > 0.70 else random.choice(OBS['sin_resultado'])

def volumen_dia(depto, fecha):
    dia  = fecha.weekday()
    vol  = VOLUMEN_BASE[depto]
    base = (vol['lun']     if dia == 0 else
            vol['mar_jue'] if dia <= 3 else
            vol['vie']     if dia == 4 else
            vol['sab']     if dia == 5 else vol['dom'])
    if base == 0: return 0
    var = int(base * 0.15)
    return random.randint(max(1, base - var), base + var)

def hora_en_turno(h, h_ini_td, h_fin_td):
    ini = int(h_ini_td.total_seconds()) // 3600
    fin = int(h_fin_td.total_seconds()) // 3600
    if ini <= fin: return ini <= h < fin
    return h >= ini or h < fin

def agente_disponible(depto, fecha, hora):
    candidatos = programacion_index.get((fecha, depto), [])
    disp = [(ag, niv) for ag, hi, hf, niv in candidatos
            if hora_en_turno(hora, hi, hf)]
    if not disp:
        disp = [(ag, niv) for ag, *_, niv in candidatos]
    if not disp:
        return None, None
    return random.choice(disp)

def generar_hora_depto(depto):
    h_ini, h_fin = VENTANA_HORARIA[depto]
    if h_ini < h_fin:
        horas = list(range(h_ini, h_fin))
        pesos = [3 if h in (9,10,11,15,16) else
                 2 if h in (8,12,14,17) else 1 for h in horas]
        hora  = random.choices(horas, weights=pesos)[0]
    else:
        hora = random.randint(0, 23)
    return hora, random.randint(0,59), random.randint(0,59)

def seleccionar_motivo(depto, tipo_es):
    tipo_bd = TIPO_A_BD.get(tipo_es, tipo_es)
    ops     = motivos_por_depto.get((depto, tipo_bd), [])
    if not ops:
        ops = motivos_por_depto.get((depto, 'inbound'), [])
    if not ops:
        ops = motivos_por_depto.get(('D4003', 'inbound'), ['NC006-SC3'])
    return random.choice(ops)

def calcular_resultado(estado, depto, nivel):
    if estado != 'atendida': return 'no_aplica'
    tasa = max(0.05, min(0.95,
               TASA_RESOLUCION.get(depto, 0.50) + AJUSTE_NIVEL.get(nivel, 0)))
    return (RESULTADO_DEPTO.get(depto, 'sin_resultado')
            if random.random() < tasa else 'sin_resultado')

def generar_duracion(estado, depto):
    if estado in ('abandonada', 'no_contacto'): return None
    rangos = {
        'D4001':(90,600),  'D4002':(120,900),
        'D4003':(60,480),  'D4004':(45,300),  'D4005':(60,420)
    }
    lo, hi = rangos.get(depto, (60,600))
    return random.randint(lo, hi)

def generar_tiempo_espera(estado, depto):
    if estado == 'abandonada': return random.randint(60, 180)
    rangos = {
        'D4001':(5,60), 'D4002':(10,90),
        'D4003':(5,75), 'D4005':(5,50)
    }
    lo, hi = rangos.get(depto, (5,90))
    return random.randint(lo, hi)

def generar_satisfaccion(resultado):
    if random.random() > 0.55: return None
    if resultado not in ('sin_resultado', 'no_aplica'):
        return random.choices([1,2,3,4,5], weights=[3,5,14,35,43])[0]
    return random.choices([1,2,3,4,5], weights=[15,22,28,22,13])[0]

# =============================================================================
# GENERACIÓN
# =============================================================================

llamadas   = []
seq_global = 1
total_ins  = 0
total_omit = 0
BATCH      = 500
deptos     = ['D4001','D4002','D4003','D4004','D4005']

fecha = FECHA_INICIO
while fecha.date() <= HOY:
    dia = fecha.weekday()
    for depto in deptos:
        if dia not in DIAS_LABORABLES[depto]:
            continue
        for _ in range(volumen_dia(depto, fecha)):
            id_ll      = f"LLA{seq_global:06d}"
            seq_global += 1

            mix  = MIX_LLAMADAS[depto]
            tipo = random.choices(list(mix), weights=list(mix.values()))[0]

            h, m, s = generar_hora_depto(depto)
            fecha_h = datetime(fecha.year, fecha.month, fecha.day, h, m, s)

            id_agente, nivel = agente_disponible(depto, fecha.date(), h)
            if id_agente is None:
                total_omit += 1
                continue

            dist   = (ESTADO_INBOUND.get(depto, ESTADO_IB_DEF)
                      if tipo == 'entrante'
                      else ESTADO_OUTBOUND.get(depto, ESTADO_OB_DEF))
            estado = random.choices(list(dist), weights=list(dist.values()))[0]

            id_motivo = seleccionar_motivo(depto, tipo)
            t_espera  = (generar_tiempo_espera(estado, depto)
                         if tipo == 'entrante' else None)
            duracion  = generar_duracion(estado, depto)

            id_dest = None
            if estado == 'transferida':
                rutas   = RUTAS_TRANSFERENCIA.get(depto,
                          [d for d in deptos if d != depto])
                d2      = random.choice(rutas)
                id_dest, _ = agente_disponible(d2, fecha.date(), h)

            resultado    = calcular_resultado(estado, depto, nivel)
            satisfaccion = (generar_satisfaccion(resultado)
                            if estado == 'atendida' else None)
            observacion  = generar_observacion(estado, resultado, depto)
            num_contacto = telefono_depto(depto)

            llamadas.append((
                id_ll, id_agente, id_motivo, tipo,
                num_contacto, fecha_h,
                t_espera, duracion, estado, id_dest,
                resultado, satisfaccion, observacion
            ))

            if len(llamadas) >= BATCH:
                cursor.executemany("""
                    INSERT INTO llamada
                    (id_llamada, id_agente, id_motivo, tipo_llamada,
                     num_contacto, fecha_hora, tiempo_espera, duracion,
                     estado, id_agente_destino, resultado_gestion,
                     satisfaccion, observaciones)
                    VALUES (%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s)
                """, llamadas)
                conexion.commit()
                total_ins += len(llamadas)
                print(f"  {total_ins:>6} insertadas | omitidas: {total_omit}")
                llamadas = []

    fecha += timedelta(days=1)

if llamadas:
    cursor.executemany("""
        INSERT INTO llamada
        (id_llamada, id_agente, id_motivo, tipo_llamada,
         num_contacto, fecha_hora, tiempo_espera, duracion,
         estado, id_agente_destino, resultado_gestion,
         satisfaccion, observaciones)
        VALUES (%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s)
    """, llamadas)
    conexion.commit()
    total_ins += len(llamadas)

# =============================================================================
# VERIFICACIÓN
# =============================================================================

print(f"\n{'='*60}")
print(f"  llamada: {total_ins:,} registros OK | omitidas: {total_omit}")
print(f"{'='*60}")

cursor.execute("SELECT COUNT(*) FROM llamada")
print(f"  Verificación MySQL             : {cursor.fetchone()[0]:,}")

cursor.execute("""
    SELECT estado, COUNT(*) FROM llamada
    GROUP BY estado ORDER BY estado
""")
print(f"\n  Por estado:")
for row in cursor.fetchall():
    print(f"    {row[0]:<14}: {row[1]:>7,}")

cursor.execute("""
    SELECT tipo_llamada, COUNT(*) FROM llamada
    GROUP BY tipo_llamada
""")
print(f"\n  Por tipo:")
for row in cursor.fetchall():
    print(f"    {row[0]:<14}: {row[1]:>7,}")

cursor.execute("""
    SELECT resultado_gestion, COUNT(*) FROM llamada
    GROUP BY resultado_gestion ORDER BY resultado_gestion
""")
print(f"\n  Por resultado_gestion:")
for row in cursor.fetchall():
    print(f"    {row[0]:<22}: {row[1]:>7,}")

cursor.execute("""
    SELECT a.id_departamento, COUNT(*) AS n
    FROM   llamada l
    JOIN   agente  a ON l.id_agente = a.id_agente
    GROUP  BY a.id_departamento
    ORDER  BY a.id_departamento
""")
print(f"\n  Por departamento:")
for row in cursor.fetchall():
    print(f"    {row[0]}: {row[1]:>7,}")

cursor.execute("""
    SELECT COUNT(*) FROM llamada
    WHERE id_agente_destino IS NOT NULL
""")
print(f"\n  Con transferencia              : {cursor.fetchone()[0]:,}")

cursor.execute("""
    SELECT COUNT(*) FROM llamada
    WHERE resultado_gestion = 'venta_cerrada'
      AND fecha_hora >= '2026-03-01'
""")
ventas_marzo = cursor.fetchone()[0]
print(f"  Ventas cerradas marzo          : {ventas_marzo:,}")
print(f"  Clientes nuevos esperados      : ~{int(ventas_marzo * 0.10):,} "
      f"(10% conversión a cliente nuevo)")
print(f"{'='*60}")
```

```python
cursor.execute("""
CREATE TABLE suscripcion (
    id_suscripcion    VARCHAR(12)   NOT NULL COMMENT 'Formato SUB-000001',
    num_documento     VARCHAR(10)   NOT NULL,
    id_servicio       VARCHAR(10)   NOT NULL,
    id_llamada_origen VARCHAR(15)   NOT NULL
                      COMMENT 'LLA: llamada real ventas marzo. HIS: histórico anterior',
    fecha_suscripcion DATE          NOT NULL,
    fecha_fin         DATE          NULL,
    estado            ENUM('activa','cancelada','suspendida','vencida')
                      NOT NULL DEFAULT 'activa',
    precio_final      DECIMAL(10,2) NOT NULL,
    CONSTRAINT chk_sus_fechas
        CHECK (fecha_fin IS NULL OR fecha_fin > fecha_suscripcion),
    PRIMARY KEY (id_suscripcion),
    FOREIGN KEY (num_documento) REFERENCES cliente(num_documento),
    FOREIGN KEY (id_servicio)   REFERENCES tercerizacion_contrato(id_servicio),
    INDEX idx_sus_cliente  (num_documento),
    INDEX idx_sus_servicio (id_servicio),
    INDEX idx_sus_estado   (estado),
    INDEX idx_sus_llamada  (id_llamada_origen),
    INDEX idx_sus_fechas   (fecha_suscripcion, fecha_fin)
) ENGINE=InnoDB DEFAULT CHARSET=utf8mb4 COLLATE=utf8mb4_unicode_ci
""")

HOY                    = date(2026, 3, 25)
FECHA_INICIO_HISTORIAL = date(2022, 2, 1)
CORTE_HIST             = date(2026, 2, 28)
RANGO_HIST             = (CORTE_HIST - FECHA_INICIO_HISTORIAL).days

cursor.execute("""
    SELECT l.id_llamada, l.num_contacto, l.fecha_hora
    FROM   llamada l JOIN agente a ON l.id_agente = a.id_agente
    WHERE  a.id_departamento   = 'D4001'
      AND  l.resultado_gestion = 'venta_cerrada'
      AND  l.estado            = 'atendida'
      AND  l.id_llamada        LIKE 'LLA%'
      AND  l.fecha_hora       >= '2026-03-01'
""")
LLAMADAS_LLA = {}
for id_ll, tel, f_hora in cursor.fetchall():
    if tel not in LLAMADAS_LLA or f_hora > LLAMADAS_LLA[tel][1]:
        LLAMADAS_LLA[tel] = (id_ll, f_hora)

tels_marzo_sus = set(LLAMADAS_LLA.keys())

cursor.execute("SELECT num_documento, telefono FROM cliente")
clientes_rows = cursor.fetchall()

cursor.execute("""
    SELECT id_servicio, fecha_lanzamiento, valor_base
    FROM   tercerizacion_contrato ORDER BY fecha_lanzamiento
""")
SERVICIOS = {}
for id_s, f_l, vb in cursor.fetchall():
    f_l = f_l.date() if isinstance(f_l, datetime) else f_l
    SERVICIOS[id_s] = {'lanzamiento': f_l, 'valor_base': float(vb)}

SERVICIOS_ORD = sorted(SERVICIOS, key=lambda s: SERVICIOS[s]['lanzamiento'])

def servicios_en(fecha):
    return [s for s in SERVICIOS_ORD if SERVICIOS[s]['lanzamiento'] <= fecha]

total_historicos = sum(1 for _, tel in clientes_rows if tel not in tels_marzo_sus)

print(f"  LLA ventas marzo               : {len(LLAMADAS_LLA):,}")
print(f"  Clientes nuevos marzo          : {sum(1 for _,t in clientes_rows if t in tels_marzo_sus):,}")
print(f"  Clientes históricos            : {total_historicos:,}")

ESTADO_X_ANTIG = {
    'antiguo':  (['activa','cancelada','suspendida','vencida'], [0.55,0.20,0.13,0.12]),
    'medio':    (['activa','cancelada','suspendida'],           [0.70,0.18,0.12]),
    'reciente': (['activa','suspendida'],                       [0.88,0.12]),
}

def estado_sus(f_ini, idx):
    meses = (HOY.year - f_ini.year)*12 + (HOY.month - f_ini.month)
    k     = 'antiguo' if meses > 12 else 'medio' if meses > 3 else 'reciente'
    ops, pesos = ESTADO_X_ANTIG[k]
    acum, bucket = 0, idx % 100
    for op, p in zip(ops, pesos):
        acum += p / sum(pesos) * 100
        if bucket < acum:
            return op
    return ops[-1]

def precio(vb, est, idx):
    descuentos = [0.00,0.05,0.10,0.15,0.20,0.25]
    d = descuentos[idx%2] if est=='cancelada' else descuentos[idx%len(descuentos)]
    return round(vb*(1-d), 2)

def fecha_fin(est, f_sus, idx):
    if est == 'activa': return None
    dias  = [45,60,90,120,180,270,365,450,540,700][idx%10]
    f_fin = f_sus + timedelta(days=dias)
    if f_fin >= HOY: f_fin = HOY - timedelta(days=5 + idx%55)
    if f_fin <= f_sus: f_fin = f_sus + timedelta(days=15)
    if f_fin >= HOY:   f_fin = f_sus + timedelta(days=15)
    return f_fin

def fecha_historica(idx, total):
    return FECHA_INICIO_HISTORIAL + timedelta(
        days=int(idx * RANGO_HIST / max(total,1)))

suscripciones   = []
sus_activas_cli = {}
seq_sus = seq_his = idx_hist = 1

for idx, (num_doc, tel) in enumerate(clientes_rows):
    es_marzo = tel in tels_marzo_sus
    if es_marzo:
        id_ll_orig, f_hora = LLAMADAS_LLA[tel]
        f_inicio = f_hora.date() if isinstance(f_hora, datetime) else f_hora
    else:
        id_ll_orig  = f"HIS{seq_his:05d}"
        seq_his    += 1
        f_inicio    = fecha_historica(idx_hist, total_historicos)
        idx_hist   += 1

    disp = servicios_en(f_inicio)
    if not disp: continue

    # Suscripción principal
    ya      = sus_activas_cli.get(num_doc, set())
    cands   = [s for s in disp if s not in ya] or disp
    id_serv = cands[idx % len(cands)]
    info    = SERVICIOS[id_serv]
    f_ini   = max(f_inicio, info['lanzamiento'])
    est     = estado_sus(f_ini, idx)
    ff      = fecha_fin(est, f_ini, idx)

    suscripciones.append((
        f"SUB-{seq_sus:06d}", num_doc, id_serv, id_ll_orig,
        f_ini, ff, est, precio(info['valor_base'], est, idx)
    ))
    if est == 'activa':
        sus_activas_cli.setdefault(num_doc, set()).add(id_serv)
    seq_sus += 1

    # Suscripciones adicionales
    meses      = (HOY.year-f_ini.year)*12 + (HOY.month-f_ini.month)
    max_extras = 2 if meses > 24 else 1 if meses > 6 else 0

    for extra in range(max_extras):
        if idx % 100 >= (55 if extra == 0 else 30): continue
        ya2    = sus_activas_cli.get(num_doc, set())
        cands2 = [s for s in disp if s not in ya2]
        if not cands2: continue
        id_s2  = cands2[(idx+extra+5) % len(cands2)]
        info2  = SERVICIOS[id_s2]
        f_ini2 = max(f_inicio + timedelta(days=30+(idx*11+extra*45)%180),
                     info2['lanzamiento'])
        if f_ini2 > HOY: continue
        est2 = estado_sus(f_ini2, idx+extra+1)
        ff2  = fecha_fin(est2, f_ini2, idx+extra)
        suscripciones.append((
            f"SUB-{seq_sus:06d}", num_doc, id_s2, id_ll_orig,
            f_ini2, ff2, est2, precio(info2['valor_base'], est2, idx+extra)
        ))
        if est2 == 'activa':
            sus_activas_cli.setdefault(num_doc, set()).add(id_s2)
        seq_sus += 1

violaciones = [s for s in suscripciones if s[5] is not None and s[5] <= s[4]]
print(f"  Suscripciones generadas        : {len(suscripciones):,}")
print(f"  Violaciones fecha_fin<=inicio  : {len(violaciones)} {'✓' if not violaciones else '✗'}")
if violaciones: raise ValueError("Violaciones detectadas.")

cursor.executemany("""
    INSERT INTO suscripcion
    (id_suscripcion, num_documento, id_servicio, id_llamada_origen,
     fecha_suscripcion, fecha_fin, estado, precio_final)
    VALUES (%s,%s,%s,%s,%s,%s,%s,%s)
""", suscripciones)
conexion.commit()

# riesgo_churn y perfil
cursor.execute("""
    UPDATE cliente c
    JOIN (
        SELECT num_documento,
               MIN(fecha_suscripcion)         AS primera_sus,
               COUNT(*)                       AS n_sus,
               SUM(estado='activa')           AS n_activas,
               SUM(estado='suspendida')       AS n_susp
        FROM suscripcion GROUP BY num_documento
    ) d ON c.num_documento = d.num_documento
    SET c.perfil = CASE
        WHEN c.tipo_documento='NIT'
             AND TIMESTAMPDIFF(MONTH,d.primera_sus,%s)>=12 THEN 'vip'
        WHEN TIMESTAMPDIFF(MONTH,d.primera_sus,%s)>=24
             AND d.n_sus>=2                                THEN 'vip'
        WHEN TIMESTAMPDIFF(MONTH,d.primera_sus,%s)>=12
             OR  d.n_sus>=2                                THEN 'alta'
        ELSE 'normal' END
""", (HOY, HOY, HOY))

cursor.execute("""
    UPDATE cliente c
    JOIN (
        SELECT num_documento,
               SUM(estado='activa')     AS n_activas,
               SUM(estado='suspendida') AS n_susp
        FROM suscripcion GROUP BY num_documento
    ) d ON c.num_documento = d.num_documento
    SET c.riesgo_churn = CASE
        WHEN d.n_activas=0 AND d.n_susp=0 THEN NULL
        WHEN d.n_activas=0 AND d.n_susp>0 THEN 'alto'
        WHEN c.perfil='vip'       THEN 'bajo'
        WHEN c.perfil='alta'      THEN 'medio'
        ELSE 'alto' END
""")
conexion.commit()

# =============================================================================
# VERIFICACIÓN
# =============================================================================

cursor.execute("SELECT COUNT(*) FROM suscripcion")
print(f"\n{'='*60}")
print(f"  suscripcion: {cursor.fetchone()[0]:,} registros OK")

cursor.execute("""
    SELECT estado, COUNT(*),
           ROUND(AVG(TIMESTAMPDIFF(MONTH,fecha_suscripcion,%s)),1) AS ant,
           ROUND(AVG(precio_final),0) AS precio
    FROM suscripcion GROUP BY estado ORDER BY COUNT(*) DESC
""", (HOY,))
print(f"\n  {'Estado':<14} {'N':>6}  {'Ant(m)':>7}  {'Precio prom':>12}")
print(f"  {'─'*44}")
for r in cursor.fetchall():
    print(f"  {r[0]:<14} {r[1]:>6}  {r[2]:>7}  {float(r[3]):>12,.0f}")

cursor.execute("""
    SELECT perfil, riesgo_churn, COUNT(*) AS n
    FROM cliente GROUP BY perfil, riesgo_churn
    ORDER BY perfil, riesgo_churn
""")
print(f"\n  {'Prioridad':<10} {'Churn':<8} {'N':>5}")
print(f"  {'─'*25}")
for r in cursor.fetchall():
    print(f"  {str(r[0]):<10} {str(r[1]):<8} {r[2]:>5}")

cursor.execute("""
    SELECT COUNT(DISTINCT num_documento) AS cli,
           COUNT(*) AS total,
           ROUND(COUNT(*)/COUNT(DISTINCT num_documento),2) AS x_cli,
           SUM(estado='activa') AS activas,
           SUM(estado='cancelada') AS canceladas,
           SUM(estado='suspendida') AS suspendidas,
           SUM(estado='vencida') AS vencidas
    FROM suscripcion
""")
r = cursor.fetchone()
print(f"\n  Clientes con suscripción       : {r[0]:,}")
print(f"  Total suscripciones            : {r[1]:,}")
print(f"  Por cliente (promedio)         : {r[2]}")
print(f"  Activas/Canceladas             : {r[3]:,}/{r[4]:,}")
print(f"  Suspendidas/Vencidas           : {r[5]:,}/{r[6]:,}")

cursor.execute("""
    SELECT MIN(fecha_suscripcion), MAX(fecha_suscripcion) FROM suscripcion
""")
r = cursor.fetchone()
print(f"  Rango fechas                   : {r[0]} → {r[1]}")

cursor.execute("""
    SELECT COUNT(*) FROM suscripcion WHERE id_llamada_origen LIKE 'LLA%'
""")
print(f"  Con origen LLA (marzo)         : {cursor.fetchone()[0]:,}")

cursor.execute("""
    SELECT COUNT(*) FROM suscripcion WHERE id_llamada_origen LIKE 'HIS%'
""")
print(f"  Con origen HIS (histórico)     : {cursor.fetchone()[0]:,}")

cursor.execute("""
    SELECT COUNT(*) FROM cliente WHERE perfil != 'normal'
""")
print(f"\n  Clientes prioridad alta/vip    : {cursor.fetchone()[0]:,}")

cursor.execute("""
    SELECT COUNT(*) FROM cliente WHERE riesgo_churn IN ('bajo','medio')
""")
print(f"  Clientes churn bajo/medio      : {cursor.fetchone()[0]:,}")
print(f"{'='*60}")
```

```python
cursor.execute("""
CREATE TABLE caso (
    id_caso             VARCHAR(12)  NOT NULL COMMENT 'Formato CAS-000001',
    id_llamada          VARCHAR(15)  NOT NULL,
    num_documento       VARCHAR(10)  NOT NULL,
    id_suscripcion      VARCHAR(12)  NOT NULL,
    tipo_caso           ENUM('soporte_tecnico','queja','reclamo',
                             'consulta','solicitud') NOT NULL,
    id_agente_asignado  VARCHAR(15)  NULL,
    descripcion         TEXT         NOT NULL,
    estado              ENUM('abierto','en_proceso','resuelto','cerrado')
                        NOT NULL DEFAULT 'abierto',
    prioridad           ENUM('baja','media','alta','critica')
                        NOT NULL DEFAULT 'media',
    medio_resolucion    ENUM('llamada','email','gestion_interna') NULL,
    sla_cumplido        TINYINT(1)   NULL,
    fecha_apertura      DATETIME     NOT NULL,
    fecha_asignacion    DATETIME     NULL,
    fecha_cierre        DATETIME     NULL,
    CONSTRAINT chk_caso_cierre
        CHECK (fecha_cierre    IS NULL OR fecha_cierre    > fecha_apertura),
    CONSTRAINT chk_caso_asignacion
        CHECK (fecha_asignacion IS NULL OR fecha_asignacion >= fecha_apertura),
    PRIMARY KEY (id_caso),
    FOREIGN KEY (id_llamada)         REFERENCES llamada(id_llamada),
    FOREIGN KEY (num_documento)      REFERENCES cliente(num_documento),
    FOREIGN KEY (id_suscripcion)     REFERENCES suscripcion(id_suscripcion),
    FOREIGN KEY (id_agente_asignado) REFERENCES agente(id_agente),
    INDEX idx_caso_cliente (num_documento),
    INDEX idx_caso_estado  (estado),
    INDEX idx_caso_tipo    (tipo_caso),
    INDEX idx_caso_fecha   (fecha_apertura),
    INDEX idx_caso_sla     (sla_cumplido)
) ENGINE=InnoDB DEFAULT CHARSET=utf8mb4 COLLATE=utf8mb4_unicode_ci
""")

TOPE = datetime(2026, 3, 25, 23, 59)

cursor.execute("""
    SELECT a.id_agente, a.id_departamento, c.categoria
    FROM   agente a JOIN cargo c ON a.id_cargo = c.id_cargo
    WHERE  a.estado = 'activo'
""")
agentes_info_caso = {r[0]: {'depto':r[1],'nivel':r[2]} for r in cursor.fetchall()}

cursor.execute("SELECT telefono, num_documento FROM cliente")
TEL_A_DOC = {r[0]: r[1] for r in cursor.fetchall()}

cursor.execute("""
    SELECT s.num_documento, s.id_suscripcion, s.id_servicio, s.estado
    FROM   suscripcion s WHERE s.estado IN ('activa','suspendida')
""")
SUS_X_DOC = {}
for doc, sus_id, serv_id, est in cursor.fetchall():
    SUS_X_DOC.setdefault(doc, []).append((sus_id, serv_id, est))

cursor.execute("""
    SELECT p.id_agente, p.fecha_turno, t.hora_inicio, t.hora_fin
    FROM   programacion p
    JOIN   turno        t ON p.id_turno  = t.id_turno
    JOIN   agente       a ON p.id_agente = a.id_agente
    WHERE  p.fecha_turno BETWEEN '2026-03-01' AND '2026-03-25'
      AND  p.asistencia  = 'asistido'
      AND  a.estado      = 'activo'
      AND  a.id_departamento IN ('D4002','D4003')
""")
prog_index_caso = {}
for id_ag, fecha, h_ini, h_fin in cursor.fetchall():
    if id_ag not in agentes_info_caso: continue
    depto_ag = agentes_info_caso[id_ag]['depto']
    prog_index_caso.setdefault((fecha, depto_ag), []).append(
        (id_ag, h_ini, h_fin)
    )

cursor.execute("""
    SELECT l.id_llamada, l.num_contacto, l.fecha_hora,
           l.duracion, l.resultado_gestion,
           a.id_departamento, l.id_motivo
    FROM   llamada l JOIN agente a ON l.id_agente = a.id_agente
    WHERE  l.estado           = 'atendida'
      AND  a.id_departamento IN ('D4002','D4003')
      AND  l.duracion        >= 90
      AND  l.fecha_hora BETWEEN '2026-03-01' AND '2026-03-25 23:59:59'
""")
llamadas_eligibles = []
for id_ll, tel, fh, dur, res, dep, id_motivo in cursor.fetchall():
    num_doc = TEL_A_DOC.get(tel)
    if num_doc and num_doc in SUS_X_DOC:
        llamadas_eligibles.append(
            (id_ll, tel, fh, dur, res, dep, num_doc, id_motivo)
        )

print(f"  Llamadas elegibles             : {len(llamadas_eligibles):,}")

cursor.execute("""
    SELECT id_agente, id_departamento FROM agente
    WHERE  estado = 'activo' AND id_departamento IN ('D4002','D4003')
""")
AGS_CASO = {}
for ag, dep in cursor.fetchall():
    AGS_CASO.setdefault(dep, []).append(ag)

MOTIVO_A_TIPO_CASO = {
    'IN2-ST001':'soporte_tecnico','IN2-ST002':'soporte_tecnico',
    'IN2-ST003':'soporte_tecnico','IN2-ST004':'soporte_tecnico',
    'IN2-ST005':'soporte_tecnico','OUT1-S001':'soporte_tecnico',
    'OUT1-S002':'soporte_tecnico','OUT1-S003':'soporte_tecnico',
    'IN1-SC004':'queja',          'IN1-SC005':'reclamo',
    'IN1-SC006':'solicitud',      'IN1-SC001':'consulta',
    'IN1-SC002':'consulta',       'IN1-SC003':'consulta',
    'IN3-V001':'consulta',        'IN3-V002':'consulta',
    'IN3-V003':'consulta',        'IN3-V004':'consulta',
    'IN4-F001':'consulta',        'IN4-F002':'reclamo',
    'IN4-F003':'consulta',        'IN4-F004':'solicitud',
    'IN4-F005':'solicitud',       'OUT4-F001':'consulta',
    'OUT4-F002':'reclamo',        'OUT4-F003':'reclamo',
    'OUT4-F004':'reclamo',        'OUT4-F005':'solicitud',
    'OUT5-T001':'consulta',       'OUT5-T002':'consulta',
    'OUT5-T003':'consulta',       'OUT5-T004':'consulta',
    'OUT5-T005':'consulta',       'OUT5-T006':'consulta',
    'OUT5-T007':'consulta',       'OUT3-V001':'consulta',
    'OUT3-V002':'consulta',       'OUT3-V003':'consulta',
    'OUT3-V004':'consulta',       'OUT3-V005':'consulta',
    'NC006-SC3':'consulta',
}

PRIO_X_TIPO = {
    ('D4002','soporte_tecnico'):[0.05,0.25,0.45,0.25],
    ('D4002','queja'):          [0.15,0.45,0.35,0.05],
    ('D4002','consulta'):       [0.50,0.40,0.10,0.00],
    ('D4003','reclamo'):        [0.05,0.30,0.45,0.20],
    ('D4003','consulta'):       [0.45,0.45,0.10,0.00],
    ('D4003','solicitud'):      [0.25,0.50,0.25,0.00],
    ('D4003','queja'):          [0.10,0.40,0.40,0.10],
}
PRIO_NIVELES = ['baja','media','alta','critica']
SLA_H        = {'baja':72,'media':24,'alta':8,'critica':2}
RANGOS_SLA   = {
    ('critica',True):(0.3,1.8),  ('alta',True):(0.5,7.5),
    ('media',True):(1.0,23.0),   ('baja',True):(2.0,70.0),
    ('critica',False):(2.5,20.0),('alta',False):(9.0,48.0),
    ('media',False):(25.0,72.0), ('baja',False):(73.0,120.0),
}

DESC_CASO = {
    ('D4002','soporte_tecnico',True):[
        "Falla técnica resuelta en llamada mediante diagnóstico remoto.",
        "Interrupción del servicio solucionada, configuración ajustada.",
        "Problema de acceso resuelto con restablecimiento de credenciales.",
        "Incidente técnico cerrado; cliente confirmó restablecimiento.",
    ],
    ('D4002','soporte_tecnico',False):[
        "Falla intermitente sin causa identificada, requiere visita técnica.",
        "Interrupción del servicio escalada a soporte nivel 2.",
        "Degradación del servicio reportada; pendiente diagnóstico avanzado.",
        "Problema de configuración complejo; derivado a especialista.",
    ],
    ('D4002','queja',True):["Queja por tiempo de respuesta gestionada satisfactoriamente.",
                            "Inconformidad técnica resuelta con compensación aplicada."],
    ('D4002','queja',False):["Queja reiterada sobre calidad del servicio, pendiente revisión.",
                             "Cliente insatisfecho con el nivel de soporte recibido."],
    ('D4002','consulta',True):["Consulta técnica resuelta satisfactoriamente."],
    ('D4002','consulta',False):["Consulta técnica compleja derivada al área."],
    ('D4003','reclamo',True):["Reclamo atendido; ajuste aplicado en la cuenta del cliente.",
                              "Cobro no reconocido verificado y revertido exitosamente."],
    ('D4003','reclamo',False):["Reclamo por cobro indebido derivado a facturación para validación.",
                               "Inconformidad con cargo adicional, pendiente respuesta del área."],
    ('D4003','consulta',True):["Consulta sobre condiciones del plan resuelta satisfactoriamente.",
                               "Información de renovación entregada al cliente.",
                               "Cambio de plan procesado correctamente."],
    ('D4003','consulta',False):["Consulta compleja derivada al área responsable.",
                                "Solicitud de información requiere gestión adicional."],
    ('D4003','solicitud',True):["Solicitud de baja procesada correctamente.",
                                "Actualización de datos del cliente completada."],
    ('D4003','solicitud',False):["Solicitud de baja en proceso, verificación de deuda pendiente.",
                                 "Trámite de modificación de plan en revisión."],
    ('D4003','queja',True):["Queja formal atendida y gestionada.",
                            "Inconformidad resuelta con compensación aplicada."],
    ('D4003','queja',False):["Queja formal registrada, pendiente respuesta.",
                             "Cliente insatisfecho con la atención recibida."],
}

def desc_caso(depto, tipo, exitoso):
    pool = DESC_CASO.get((depto, tipo, exitoso))
    if not pool:
        fallback = ('D4002','soporte_tecnico',exitoso) if depto=='D4002' \
                   else ('D4003','consulta',exitoso)
        pool = DESC_CASO.get(fallback, ["Caso registrado para seguimiento."])
    return random.choice(pool)

def hora_en_turno(h, h_ini_td, h_fin_td):
    ini = int(h_ini_td.total_seconds()) // 3600
    fin = int(h_fin_td.total_seconds()) // 3600
    if ini <= fin: return ini <= h < fin
    return h >= ini or h < fin

def agente_caso(depto, fecha_dt):
    fecha     = fecha_dt.date()
    hora      = fecha_dt.hour
    candidatos = prog_index_caso.get((fecha, depto), [])
    disp = [ag for ag, hi, hf in candidatos if hora_en_turno(hora, hi, hf)]
    if not disp:
        disp = [ag for ag, *_ in candidatos]
    return random.choice(disp) if disp else None

def elegir_sus(num_doc, tipo_caso):
    pool = SUS_X_DOC.get(num_doc, [])
    if not pool: return None, None
    if tipo_caso in ('soporte_tecnico','solicitud'):
        cands = [(s,sv) for s,sv,e in pool if e=='suspendida'] or \
                [(s,sv) for s,sv,e in pool if e=='activa']
    elif tipo_caso == 'reclamo':
        cands = [(s,sv) for s,sv,e in pool]
    else:
        cands = [(s,sv) for s,sv,e in pool if e=='activa'] or \
                [(s,sv) for s,sv,e in pool]
    return random.choice(cands) if cands else (None, None)

def calcular_fechas(f_ap, estado, prioridad):
    f_asig = f_cie = sla = None
    if estado not in ('en_proceso','resuelto','cerrado'):
        return f_asig, f_cie, sla
    f_asig = f_ap + timedelta(minutes=random.randint(5,90))
    if f_asig > TOPE: f_asig = TOPE
    if estado not in ('resuelto','cerrado'):
        return f_asig, f_cie, sla
    sla_ok       = random.random() < 0.65
    h_min, h_max = RANGOS_SLA[(prioridad, sla_ok)]
    f_cie        = f_ap + timedelta(hours=random.uniform(h_min, h_max))
    if f_cie > TOPE: f_cie = TOPE - timedelta(minutes=random.randint(10,60))
    if f_cie <= f_ap: f_cie = f_ap + timedelta(minutes=30)
    if f_cie > TOPE:  f_cie = TOPE
    horas = (f_cie - f_ap).total_seconds() / 3600
    sla   = 1 if horas <= SLA_H[prioridad] else 0
    return f_asig, f_cie, sla

# =============================================================================
# PROBABILIDAD DE GENERACIÓN DE CASO
# En un BPO real no toda llamada atendida genera caso de seguimiento
# Solo el 40% requiere gestión posterior documentada
# =============================================================================

# Probabilidad diferenciada por resultado — más coherente con la realidad
PROB_CASO_X_RESULTADO = {
    'ticket_resuelto': 0.55,  # soporte resuelto en llamada → seguimiento moderado
    'caso_resuelto':   0.45,  # servicio resuelto → menos seguimiento
    'sin_resultado':   0.70,  # no resuelto → casi siempre genera caso abierto
    '__default__':     0.30,  # resto → bajo seguimiento
}

llamadas_usadas = set()
casos = []
seq   = 1
omitidas_prob = 0

for id_ll, tel, f_hora, dur, resultado, depto, num_doc, id_motivo \
        in llamadas_eligibles:

    if id_ll in llamadas_usadas: continue

    # Filtro probabilístico diferenciado por resultado
    prob = PROB_CASO_X_RESULTADO.get(resultado,
           PROB_CASO_X_RESULTADO['__default__'])
    if random.random() > prob:
        omitidas_prob += 1
        continue

    tipo_caso = MOTIVO_A_TIPO_CASO.get(id_motivo, 'consulta')
    id_sus, _ = elegir_sus(num_doc, tipo_caso)
    if id_sus is None: continue

    pesos_prio = PRIO_X_TIPO.get((depto, tipo_caso), [0.20,0.45,0.30,0.05])
    prioridad  = random.choices(PRIO_NIVELES, weights=pesos_prio)[0]

    if resultado in ('ticket_resuelto','caso_resuelto'):
        estado = random.choices(['resuelto','cerrado'], weights=[0.40,0.60])[0]
    elif resultado == 'sin_resultado':
        estado = random.choices(['abierto','en_proceso'], weights=[0.45,0.55])[0]
    else:
        estado = random.choices(
            ['abierto','en_proceso','resuelto','cerrado'],
            weights=[0.15,0.25,0.30,0.30])[0]

    if prioridad in ('critica','alta') and estado == 'abierto':
        estado = 'en_proceso'

    f_ap   = f_hora if isinstance(f_hora, datetime) \
             else datetime.combine(f_hora, datetime.min.time())
    f_asig, f_cie, sla = calcular_fechas(f_ap, estado, prioridad)

    ag_asig = None
    if estado in ('en_proceso','resuelto','cerrado'):
        ag_asig = agente_caso(depto, f_ap)

    medio = None
    if estado in ('resuelto','cerrado') and f_cie:
        medio = random.choices(
            ['llamada','email','gestion_interna'],
            weights=[0.55,0.25,0.20] if tipo_caso in ('soporte_tecnico','queja')
                    else [0.40,0.35,0.25])[0]

    casos.append((
        f"CAS-{seq:06d}",
        id_ll, num_doc, id_sus,
        tipo_caso, ag_asig,
        desc_caso(depto, tipo_caso, estado in ('resuelto','cerrado')),
        estado, prioridad, medio, sla,
        f_ap, f_asig, f_cie
    ))
    llamadas_usadas.add(id_ll)
    seq += 1

# =============================================================================
# VERIFICACIÓN PRE-INSERT
# =============================================================================

nulos  = [sum(1 for c in casos if c[i] is None) for i in [1,2,3]]
viols  = [c for c in casos if c[13] is not None and c[13] <= c[11]]
viols2 = [c for c in casos if c[12] is not None and c[12] <  c[11]]

print(f"\n  Llamadas elegibles             : {len(llamadas_eligibles):,}")
print(f"  Omitidas por probabilidad      : {omitidas_prob:,}")
print(f"  Casos generados                : {len(casos):,}")
print(f"  Tasa generación caso           : "
      f"{round(len(casos)/len(llamadas_eligibles)*100,1)}%")
print(f"  id_llamada NULL                : {nulos[0]}  {'✓' if nulos[0]==0 else '✗'}")
print(f"  num_documento NULL             : {nulos[1]}  {'✓' if nulos[1]==0 else '✗'}")
print(f"  id_suscripcion NULL            : {nulos[2]}  {'✓' if nulos[2]==0 else '✗'}")
print(f"  Violaciones fecha_cierre       : {len(viols)}  {'✓' if not viols else '✗'}")
print(f"  Violaciones fecha_asignacion   : {len(viols2)}  {'✓' if not viols2 else '✗'}")

if any([nulos[0],nulos[1],nulos[2],viols,viols2]):
    raise ValueError("Violaciones de trazabilidad. Revisar antes de insertar.")

cursor.executemany("""
    INSERT INTO caso
    (id_caso, id_llamada, num_documento, id_suscripcion,
     tipo_caso, id_agente_asignado, descripcion,
     estado, prioridad, medio_resolucion, sla_cumplido,
     fecha_apertura, fecha_asignacion, fecha_cierre)
    VALUES (%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s)
""", casos)
conexion.commit()

# =============================================================================
# VERIFICACIÓN FINAL
# =============================================================================

from collections import Counter
cerrados = [c for c in casos if c[10] is not None]
sla_ok   = sum(1 for c in cerrados if c[10] == 1)

print(f"\n{'='*60}")
print(f"  caso: {len(casos):,} registros OK")

print(f"\n  Por estado:")
for k,v in sorted(Counter(c[7] for c in casos).items()):
    print(f"    {k:<14}: {v:>6,}")

print(f"\n  Por tipo_caso:")
for k,v in sorted(Counter(c[4] for c in casos).items()):
    print(f"    {k:<22}: {v:>6,}")

print(f"\n  Por prioridad:")
for k,v in sorted(Counter(c[8] for c in casos).items()):
    print(f"    {k:<14}: {v:>6,}")

cursor.execute("""
    SELECT a.id_departamento, d.nombre_dpto, COUNT(*) AS n
    FROM   caso ca
    JOIN   llamada     l ON ca.id_llamada      = l.id_llamada
    JOIN   agente      a ON l.id_agente        = a.id_agente
    JOIN   departamento d ON a.id_departamento = d.id_departamento
    GROUP  BY a.id_departamento, d.nombre_dpto
    ORDER  BY a.id_departamento
""")
print(f"\n  Por departamento:")
for r in cursor.fetchall():
    print(f"    {r[0]} {r[1]:<25}: {r[2]:>6,}")

if cerrados:
    print(f"\n  SLA cumplido                   : {sla_ok:,}/{len(cerrados):,} "
          f"({round(sla_ok/len(cerrados)*100,1)}%)")

print(f"\n  Omitidas por probabilidad      : {omitidas_prob:,}")
print(f"  Tasa real generación caso      : "
      f"{round(len(casos)/len(llamadas_eligibles)*100,1)}%")
print(f"  Casos por día (promedio)       : {round(len(casos)/25,0):.0f}")
print(f"{'='*60}")
```

SLA - Acuerdo de Nivel de Servicio
Tiempo máximo comprometido para resolver o responder un caso según su tipo y prioridad. Ejemplo: una queja de prioridad alta debe cerrarse en máximo 4 horas, un soporte técnico en 24 horas, etc.
En este caso:

Cumplido (1) : 4,554 → casos cerrados dentro del tiempo acordado
Fallido (0) : 2,334 → casos que superaron el tiempo límite sin resolverse
Sin cerrar : 7,508 → casos aún abiertos, todavía no aplica el resultado

```python
# FACTURA
cursor.execute("""
CREATE TABLE factura (
    id_factura      VARCHAR(12)   NOT NULL
                    COMMENT 'Formato FAC-000001',
    id_suscripcion  VARCHAR(12)   NOT NULL
                    COMMENT 'Suscripciones de marzo 2026 que originan esta factura',
    num_documento   VARCHAR(10)   NOT NULL
                    COMMENT 'Cliente facturado',
    id_servicio     VARCHAR(10)   NOT NULL
                    COMMENT 'Servicio adquirido',
    fecha_emision   DATE          NOT NULL
                    COMMENT 'Igual a suscripcion.fecha_suscripcion (fecha de la llamada de venta LLA)',
    fecha_pago      DATE          NULL
                    COMMENT 'fecha_emision + 30 dias si el cliente pago. NULL si pendiente (mayoria vence en abril 2026).',
    valor_base      DECIMAL(12,2) NOT NULL
                    COMMENT 'Igual a suscripcion.precio_final',
    iva             DECIMAL(12,2) NOT NULL DEFAULT 0.00
                    COMMENT 'IVA 19% sobre valor_base para servicios gravados. 0 para streaming, educacion y noticias.',
    descuento       DECIMAL(12,2) NOT NULL DEFAULT 0.00
                    COMMENT 'Descuento por fidelidad segun suscripciones vigentes: 0% (1 sus) | 5% (2 sus) | 10% (3+ sus)',
    valor_total     DECIMAL(12,2) NOT NULL
                    COMMENT 'valor_base + iva - descuento',
    CONSTRAINT chk_fac_pago
        CHECK (fecha_pago IS NULL OR fecha_pago >= fecha_emision),
 
    PRIMARY KEY (id_factura),
    UNIQUE  KEY uq_fac_suscripcion (id_suscripcion),
 
    FOREIGN KEY (id_suscripcion)  REFERENCES suscripcion(id_suscripcion),
    FOREIGN KEY (num_documento)   REFERENCES cliente(num_documento),
    FOREIGN KEY (id_servicio)     REFERENCES tercerizacion_contrato(id_servicio),
 
    INDEX idx_fac_cliente  (num_documento),
    INDEX idx_fac_servicio (id_servicio),
    INDEX idx_fac_emision  (fecha_emision),
    INDEX idx_fac_pago     (fecha_pago)
 
) ENGINE=InnoDB DEFAULT CHARSET=utf8mb4 COLLATE=utf8mb4_unicode_ci
""")
print(" Tabla factura creada")
```

```python
# Solo suscripciones de marzo 2026 (id_llamada_origen = LLA real)
cursor.execute("""
    SELECT s.id_suscripcion,
           s.num_documento,
           s.id_servicio,
           s.estado            AS estado_sus,
           s.precio_final,
           s.fecha_suscripcion,
           l.fecha_hora        AS fecha_llamada
    FROM   suscripcion s
    LEFT JOIN llamada l ON s.id_llamada_origen = l.id_llamada
    WHERE  s.id_llamada_origen LIKE 'LLA%'
    ORDER  BY s.id_suscripcion
""")
suscripciones_marzo = cursor.fetchall()
print(f"  Suscripciones LLA de marzo     : {len(suscripciones_marzo)}")

if not suscripciones_marzo:
    raise ValueError(
        "No se encontraron suscripciones LLA. "
        "Verificar que el script de suscripciones fue ejecutado."
    )

# Categoría de empresa por servicio → exención de IVA
cursor.execute("""
    SELECT tc.id_servicio, e.categoria
    FROM   tercerizacion_contrato tc
    JOIN   empresa e ON tc.id_empresa = e.nit
""")
CAT_SERV = {row[0]: row[1] for row in cursor.fetchall()}
EXENTOS  = {'streaming', 'educacion', 'noticias'}

def calcular_iva(id_serv, base):
    """IVA 19% sobre valor_base. 0 para categorías exentas."""
    return 0.00 if CAT_SERV.get(id_serv) in EXENTOS else round(float(base) * 0.19, 2)

# Suscripciones vigentes HOY por cliente (activas + suspendidas)
cursor.execute("""
    SELECT num_documento, COUNT(*) AS n_vigentes
    FROM   suscripcion
    WHERE  estado IN ('activa', 'suspendida')
    GROUP  BY num_documento
""")
N_VIG_CLI = {row[0]: row[1] for row in cursor.fetchall()}

# =============================================================================
# GENERACIÓN
# =============================================================================

facturas = []
seq      = 1

for (id_sus, num_doc, id_serv, estado_sus,
     precio_raw, fecha_sus, fecha_llamada) in suscripciones_marzo:

    precio = float(precio_raw) if precio_raw is not None else 0.0

    # fecha_emision: fecha de la llamada LLA; fallback a fecha_suscripcion
    if fecha_llamada is not None:
        fecha_emision = (fecha_llamada.date()
                         if isinstance(fecha_llamada, datetime)
                         else fecha_llamada)
    elif fecha_sus is not None:
        fecha_emision = (fecha_sus if isinstance(fecha_sus, date)
                         else date.fromisoformat(str(fecha_sus)))
    else:
        fecha_emision = HOY

    # ── IVA ──────────────────────────────────────────────────────────────────
    iva = calcular_iva(id_serv, precio)

    # ── Descuento por fidelidad ───────────────────────────────────────────────
    n_vig = N_VIG_CLI.get(num_doc, 1)
    desc  = (round(precio * 0.10, 2) if n_vig >= 3 else
             round(precio * 0.05, 2) if n_vig == 2 else 0.00)

    total = round(precio + iva - desc, 2)

    # ── Probabilidad de pago ──────────────────────────────────────────────────
    # Vencimiento = fecha_emision + 30 días
    # Casi todos los contratos de marzo vencen en abril → prob_pago baja
    vencimiento = fecha_emision + timedelta(days=30)

    if vencimiento <= HOY:
        prob_pago = 0.70 if estado_sus in ('activa', 'suspendida') else 0.50
    else:
        prob_pago = 0.15

    f_pago = None
    if random.random() < prob_pago:
        f_pago = min(vencimiento, HOY)

    facturas.append((
        f"FAC-{seq:06d}",
        id_sus,
        num_doc,
        id_serv,
        fecha_emision,
        f_pago,
        precio,
        iva,
        desc,
        total
    ))
    seq += 1

# =============================================================================
# VERIFICACIÓN PRE-INSERT
# =============================================================================

viol_pago = [f for f in facturas if f[5] is not None and f[5] < f[4]]
print(f"\n  Facturas generadas             : {len(facturas)}")
print(f"  Violaciones fecha_pago<emision : {len(viol_pago)}  "
      f"{'✓ OK' if not viol_pago else '✗ REVISAR'}")

if viol_pago:
    for v in viol_pago[:3]:
        print(f"    id_sus={v[1]}  emision={v[4]}  pago={v[5]}")
    raise ValueError(f"{len(viol_pago)} violaciones del CHECK antes de insertar.")

# =============================================================================
# INSERCIÓN
# =============================================================================

insertadas = 0
errores    = []

for idx, fila in enumerate(facturas):
    try:
        cursor.execute("""
            INSERT INTO factura
                (id_factura, id_suscripcion, num_documento, id_servicio,
                 fecha_emision, fecha_pago,
                 valor_base, iva, descuento, valor_total)
            VALUES (%s,%s,%s,%s,%s,%s,%s,%s,%s,%s)
        """, fila)
        conexion.commit()
        insertadas += 1
    except Exception as e:
        conexion.rollback()
        errores.append({'idx': idx, 'error': str(e), 'fila': fila})

# =============================================================================
# RESUMEN
# =============================================================================

print(f"\n  Insertadas correctamente       : {insertadas}")
print(f"  Filas con error                : {len(errores)}")

if errores:
    COLS = ['id_factura','id_suscripcion','num_documento','id_servicio',
            'fecha_emision','fecha_pago','valor_base','iva','descuento','valor_total']
    print(f"\n  ── Detalle de errores (primeros 5) ──")
    for err in errores[:5]:
        print(f"\n  [{err['idx']}] {err['error']}")
        for col, val in zip(COLS, err['fila']):
            print(f"    {col:<18}: {type(val).__name__:<12} = {val}")
else:
    pagadas    = sum(1 for f in facturas if f[5] is not None)
    pendientes = insertadas - pagadas
    con_desc   = sum(1 for f in facturas if f[8] > 0)
    exentas    = sum(1 for f in facturas if f[7] == 0.0)

    total_base = sum(f[6]  for f in facturas)
    total_iva  = sum(f[7]  for f in facturas)
    total_desc = sum(f[8]  for f in facturas)
    total_tot  = sum(f[9]  for f in facturas)

    print(f"\n  ── Desglose facturas de marzo 2026 ──")
    print(f"  Pagadas (fecha_pago poblada)  : {pagadas}  "
          f"({round(pagadas/insertadas*100,1) if insertadas else 0}%)")
    print(f"  Pendientes (vencen en abril)  : {pendientes}  "
          f"({round(pendientes/insertadas*100,1) if insertadas else 0}%)")
    print(f"  Con descuento fidelidad       : {con_desc}  "
          f"({round(con_desc/insertadas*100,1) if insertadas else 0}%)")
    print(f"  Exentas de IVA                : {exentas}  "
          f"({round(exentas/insertadas*100,1) if insertadas else 0}%)")

    print(f"\n  ── Valores facturados ──")
    print(f"  Suma valor_base               : ${total_base:>14,.2f}")
    print(f"  Suma IVA                      : ${total_iva:>14,.2f}")
    print(f"  Suma descuentos               : ${total_desc:>14,.2f}")
    print(f"  Suma valor_total              : ${total_tot:>14,.2f}")

    # Detalle por servicio
    cursor.execute("""
        SELECT tc.nombre_servicio,
               COUNT(*)                        AS n_facturas,
               ROUND(SUM(f.valor_base), 2)     AS base_total,
               ROUND(AVG(f.descuento), 2)      AS desc_prom,
               SUM(f.fecha_pago IS NOT NULL)   AS pagadas
        FROM   factura f
        JOIN   tercerizacion_contrato tc ON f.id_servicio = tc.id_servicio
        GROUP  BY tc.nombre_servicio
        ORDER  BY n_facturas DESC
    """)
    rows = cursor.fetchall()
    if rows:
        print(f"\n  {'Servicio':<35} {'Fac':>4} {'Base $':>14} "
              f"{'Desc prom':>10} {'Pagadas':>8}")
        print(f"  {'-'*75}")
        for r in rows:
            print(f"  {r[0]:<35} {r[1]:>4} {float(r[2]):>14,.2f} "
                  f"{float(r[3]):>10,.2f} {r[4]:>8}")
```
